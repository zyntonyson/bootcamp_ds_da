### 🧪 Ejercicio 1: Imprimir los datos en un formato amigable

Crea una función  para imprimir los detalles de cada videojuego.

Formato esperado:
```
🎮 Título: Hades
📅 Año: 2020
🕹️ Plataforma: PC
📚 Género: Roguelike
------------------------------
```


````python
def mostrar_juegos(juegos):
  for juego in juegos:
    print(f"🎮 Título: {juego}")
    print(f"📅 Año: {juegos[juego]['año']}")
    print(f"🕹️ Plataforma: {juegos[juego]['plataforma']}")
    print(f"📚 Género: {juegos[juego]['género']}")
    print()
    print("------------------------------")
````

````python
def mostrar_juegos(juegos):
  for juego, datos_juego in juegos.items():
    print(f"🎮 Título: {juego}")
    print(f"📅 Año: {datos_juego['año']}")
    print(f"🕹️ Plataforma: {datos_juego['plataforma']}")
    print(f"📚 Género: {datos_juego['género']}")
    print()
    print("------------------------------")
````

````python
def mostrar_juegos(juegos):
  for juego, datos_juego in juegos.items():
    print(juego)
    print(datos_juego)
    print()
    print("------------------------------")
````

### 🧪 Ejercicio 2: Agregar un nuevo videojuego con una función

Crea una función `agregar_videojuego(diccionario, titulo, plataforma, año, genero)` que agregue un nuevo videojuego al diccionario.

Luego, usa `len(videojuegos)` para verificar cuántos juegos hay en total.


````python
def agregar_videojuego(juegos, titulo, plataforma, año, genero):
  juego_nuevo = {'plataforma':plataforma,'año':año,'género':genero}
  juegos[titulo] = juego_nuevo
  return juegos

# Ejemplo de uso
videojuegos=agregar_videojuego(videojuegos, "Hollow Knight", "PC", 2017, "Metroidvania")


print(f"Total de videojuegos: {len(videojuegos)}")
````
